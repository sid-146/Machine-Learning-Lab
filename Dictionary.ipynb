```python
# Demonstrate application of Dictionary by taking birthday month as an input and count number of birthdays in particular month by taking month as key. Identify maximum number and minimum number of birthday.
Dic = {
    "Jan" : 0,
    "Feb" : 0,
    "Mar" : 0,
    "Apr" : 0,
    "May" : 0,
    "June" : 0,
    "July" : 0,
    "Aug" : 0,
    "Sept" : 0,
    "Oct" : 0,
    "Nov" : 0,
    "Dec" : 0
}
print(Dic)
n = int(input("Enter total number of students: "))
while n > 0:
  mon = input("Enter Your Birthday Month: ")
  Dic[mon] += 1
  n -= 1
print(Dic)

maximum = max(Dic.keys(), key = (lambda k: Dic[k]))
minimum = min(Dic.keys(), key = (lambda k: Dic[k]))

print('Maximum Number of Birthdays: ', Dic[maximum])
print('Minimum Number of Birthdays: ', Dic[minimum])
```
