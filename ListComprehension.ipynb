```python
# Print square of natural number (1,10)
L = [i*i for i in range(1,11)]
print(L)
```

```python
# Print multiplication table of 3
L = [3*i for i in range(1, 11)]
print(L)
```

```python
# Print even numbers between (0, 10)
L = [i for i in range(11) if i % 2 == 0]
print(L)
```

```python
# Count number of 6 in a number
n = int(input("Enter a number: "))
n1 = str(n)
L = [i for i in n1 if i == '6']
print(L)
print(len(L))
```

```python
# Count number of spaces in string
n = input("Enter a string: ")
L = [i for i in n if i == ' ']
print(len(L))
```

```python
# Find numbers between 1-1000 divisible by 2 and 5
L = [i for i in range(1, 1001) if i % 2 == 0 and i % 5 == 0]
print(L)
print(len(L))
```
