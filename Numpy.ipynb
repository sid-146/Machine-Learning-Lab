```python
import numpy as np
```

```python
arr = np.array([[[1, 2, 3, 9], [3, 4, 5, 10], [6, 7, 8, 10]], [[1, 2, 3, 10], [3, 4, 5, 10], [6, 7, 8, 10]], [[1, 2, 3, 10], [3, 4, 5, 10], [6, 7, 8, 10]]])
print(arr)    #prints arr
print(arr.shape)  #prints (x, y, z) dimensions
print(arr.ndim)   #prints dimension of matrix
```

```python
arr = np.array([[[1, 2, 3, 9], [3, 4, 5, 10], [6, 7, 8, 10]], [[1, 2, 3, 10], [3, 4, 5, 10], [6, 7, 8, 10]], [[1, 2, 3, 10], [3, 4, 5, 10], [6, 7, 8, 10]]])
arr.resize(3, 9)   #resize changes original array
print(arr)    
print(arr.shape)
```

```python
arr = np.array([[[1, 2, 3, 9], [3, 4, 5, 10], [6, 7, 8, 10]], [[1, 2, 3, 10], [3, 4, 5, 10], [6, 7, 8, 10]], [[1, 2, 3, 10], [3, 4, 5, 10], [6, 7, 8, 10]]])
print(arr.reshape(9, 4)) #reshape does not change original array
print(arr.shape)
```
